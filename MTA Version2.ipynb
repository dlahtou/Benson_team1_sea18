```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('mta.csv', delimiter = '\t', index_col = 0)
```

```python
df1.head(10)
```

```python
df2 = (df1.sort_values(by=['STATION', 'C/A', 'SCP', 'LINENAME', 'UNIT'])
          .groupby(['STATION', 'LINENAME'])['ENTRY_DIFFS', 'EXIT_DIFFS']
          .sum())
```

```python
df2['TOTAL DIFF'] = df2['ENTRY_DIFFS'] + df2['EXIT_DIFFS']
```
